Data Cleaning & Visualization with Raw Dataset 

Load and Explore the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the dataset data.csv using Pandas

```python
df = pd.read_csv("dataset/data.csv")
```

Display the first 10 rows using .head()

```python
print("First 10 rows of dataset:")
df.head(10)
```

Use .info() and .describe() to understand the dataset structure and summary statistics

```python
df.info()
```

```python
df.describe()
```

Visualize Unclean Data

Plot a bar chart showing Duration values for the first 20 records

```python
plt.figure(figsize=(10,5))
plt.bar(df.index[:20], df['Duration'][:20])
plt.title("Duration for First 20 Records (Unclean Data)")
plt.xlabel("Record Index")
plt.ylabel("Duration (minutes)")
plt.show()
```

Plot a line chart showing how Calories change over records

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Calories"], marker='o')
plt.title("Calories Over Records (Unclean Data)")
plt.xlabel("Record Index")
plt.ylabel("Calories")
plt.show()
```

Clean the Data

Handle missing values: Fill missing values in Calories with the mean

```python
df['Calories'] = df['Calories'].fillna(df['Calories'].mean())
```

Fix invalid data: Replace values in Duration above 120 minutes with the median

```python
median_duration = df['Duration'].median()
df.loc[df['Duration'] > 120, 'Duration'] = median_duration
```

Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
print("Cleaned Dataset (First 10 rows):")
df.head(10)
```

Visualize Cleaned Data

Recreate the same plots using the cleaned dataset:
    Bar chart of Duration for first 20 record

```python
plt.figure(figsize=(10,5))
plt.bar(df.index[:20], df['Duration'][:20])
plt.title("Duration for First 20 Records (Cleaned Data)")
plt.xlabel("Record Index")
plt.ylabel("Duration (minutes)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Calories"], marker='o')
plt.title("Calories Over Time (Cleaned Data)")
plt.xlabel("Date")
plt.ylabel("Calories")
plt.xticks(rotation=45)
plt.show()
```
